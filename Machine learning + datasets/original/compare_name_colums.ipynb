```python
import pandas as pd

file1 = "busson20.csv"
file2 = "data.csv"

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

cols1 = list(df1.columns)
cols2 = list(df2.columns)

if cols1 == cols2:
    print("✅ Hai file có cùng tên cột và cùng thứ tự")
else:
    print("❌ Hai file KHÔNG giống nhau về tên hoặc thứ tự cột")

    for i, (c1, c2) in enumerate(zip(cols1, cols2)):
        if c1 != c2:
            print(f"Khác tại vị trí {i}: file1='{c1}', file2='{c2}'")

    if len(cols1) != len(cols2):
        print("⚠️ Số lượng cột khác nhau")
```
