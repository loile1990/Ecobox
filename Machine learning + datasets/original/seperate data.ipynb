```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load CSV
df = pd.read_csv("test4G.csv")

# Column that contains the class labels
LABEL_COL = "DeviceScreenState"   # 🔁 change this to your column name

# Stratified split
train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    shuffle=True,
    stratify=df[LABEL_COL]
)

# Save to CSV files
train_df.to_csv("train.csv", index=False)
val_df.to_csv("validation.csv", index=False)

print("Files saved:")
print(" - train.csv")
print(" - validation.csv")
```
