```python
import pandas as pd

df1 = pd.read_csv("busson20.csv")
df2 = pd.read_csv("data.csv")
df3 = pd.read_csv("Ryadh20.csv")

df_union = (
    pd.concat([df1, df2, df3], ignore_index=True)
      .drop_duplicates()
)
```

```python
df_union.to_csv("union_output.csv", index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load CSV
df = pd.read_csv("union_output.csv")

# Column that contains the class labels
LABEL_COL = "DeviceScreenState"   # 🔁 change this to your column name

# Stratified split
train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    shuffle=True,
    stratify=df[LABEL_COL]
)

# Save to CSV files
train_df.to_csv("train.csv", index=False)
val_df.to_csv("validation.csv", index=False)

print("Files saved:")
print(" - train.csv")
print(" - validation.csv")
```
